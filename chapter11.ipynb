# K-mean

```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
%matplotlib inline

X, y = make_blobs(n_samples=150, 
                  n_features=2, 
                  centers=3, 
                  cluster_std=0.5, 
                  shuffle=True, 
                  random_state=0)


plt.scatter(X[:, 0], X[:, 1], c="black", marker='o', s=50)
plt.grid()
plt.tight_layout()
#plt.savefig('./figures/spheres.png', dpi=300)
plt.show()
```

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters=3,
           init="random",
           n_init=20,
           max_iter=300,
           tol=1e-4,
           random_state=9)
y_km = km.fit_predict(X)
```

```python
plt.figure(figsize=(10,10))
plt.scatter(X[y_km==0,0],
            X[y_km ==0,1],
            s=50,
            c='lightgreen',
            marker='s',
            label='cluster 1')
plt.scatter(X[y_km ==1,0],
            X[y_km ==1,1],
            s=50,
            c='orange',
            marker='o',
            label='cluster 2')
plt.scatter(X[y_km ==2,0],
            X[y_km ==2,1],
            s=50,
            c='lightblue',
            marker='v',
            label='cluster 3')
plt.scatter(km.cluster_centers_[:,0],
            km.cluster_centers_[:,1],
            s=250,
            marker='*',
            c='red',
            label='centroids')
plt.legend()
plt.grid()
plt.show()
```

###### Hard versus soft clustering
## Using the elbow method to find the optimal number of clusters

```python
print('Distortion: %.2f' % km.inertia_)
```

```python
distortions = []
for i in range(1, 11):
    km = KMeans(n_clusters=i,
           init="random",
           n_init=10,
           max_iter=300,
           tol=1e-4,
           random_state=9)
    km.fit(X)
    distortions.append(km.inertia_)
plt.figure(figsize=(10,10))
plt.plot(list(range(1,11)), distortions, marker="o")
```

# Quantifying the quality of clustering via silhouette plots

```python
km = KMeans(n_clusters=3,
           init="random",
           n_init=10,
           max_iter=300,
           tol=1e-4,
           random_state=9)
y_km = km.fit_predict(X)

from matplotlib import cm
import numpy as np
from sklearn.metrics import silhouette_samples

cluster_labels = np.unique(y_km)
n_clusters = cluster_labels.shape[0]
silhoutte_vals = silhouette_samples(X, y_km, metric="euclidean")
y_ax_lower, y_ax_upper = 0, 0
yticks = []
for i, c in enumerate(cluster_labels):
    c_silhouette = silhoutte_vals[y_km == c]
    c_silhouette.sort()
    y_ax_upper += len(c_silhouette)
    color = cm.jet(i / n_clusters)
    plt.barh(range(y_ax_lower, y_ax_upper), c_silhouette,
            height=1.0, edgecolor="none", color=color)
    yticks.append((y_ax_lower + y_ax_upper) / 2)
    y_ax_lower += len(c_silhouette)
    silhouette_avg = np.mean(silhoutte_vals)
plt.axvline(silhouette_avg,
            color="red",
            linestyle="--"
             )
plt.yticks(yticks, cluster_labels + 1)
plt.show()
```

# Organizing clusters as a hierarchical tree

## agglomerative hierarchical cluster

### complete linkage

```python
import pandas as pd
import numpy as np
np.random.seed(123)
variables = ["X", "Y", "Z"]
labels = ['ID_0','ID_1','ID_2','ID_3','ID_4']
X = np.random.random_sample([5,3]) * 10
df = pd.DataFrame(X, columns=variables, index=labels)
df
```

```python
from scipy.spatial.distance import pdist, squareform
row_dist = pd.DataFrame(squareform(
    pdist(df, metric="euclidean")),
    columns=labels, index=labels)
row_dist
```

```python
from scipy.cluster.hierarchy import linkage
help(linkage)
```

```python
from scipy.cluster.hierarchy import linkage
row_clusters1 = linkage(row_dist, method="complete", metric="euclidean")
```

```python
row_clusters2 = linkage(df.values,
                       method="complete",
                       metric="euclidean")
```

```python
pd.DataFrame(row_clusters1,
            columns=["row labels 1",
                     "row labels 2",
                     "distance",
                     "no. of items in clust"],
            index=["cluster %d" %(i+1) for i in range(row_clusters1.shape[0])])
```

```python
pd.DataFrame(row_clusters2,
            columns=["row labels 1",
                     "row labels 2",
                     "distance",
                     "no. of items in clust"],
            index=["cluster %d" %(i+1) for i in range(row_clusters2.shape[0])])
```

```python
from scipy.cluster.hierarchy import dendrogram
row_dendr = dendrogram(row_clusters2,
                      labels=labels)
plt.tight_layout()
plt.ylabel("Euclidean Distance")
```

### Attaching dendrograms to a heat map

```python
# plot row dendrogram
fig = plt.figure(figsize=(8, 8), facecolor='white')
axd = fig.add_axes([0.09, 0.1, 0.2, 0.6])

# note: for matplotlib < v1.5.1, please use orientation='right'
row_dendr = dendrogram(row_clusters2, orientation='left')

# reorder data with respect to clustering
df_rowclust = df.ix[row_dendr['leaves'][::-1]]

axd.set_xticks([])
axd.set_yticks([])

# remove axes spines from dendrogram
for i in axd.spines.values():
        i.set_visible(False)

# plot heatmap
axm = fig.add_axes([0.23, 0.1, 0.6, 0.6])  # x-pos, y-pos, width, height
cax = axm.matshow(df_rowclust, interpolation='nearest', cmap='hot_r')
fig.colorbar(cax)
axm.set_xticklabels([''] + list(df_rowclust.columns))
axm.set_yticklabels([''] + list(df_rowclust.index))

# plt.savefig('./figures/heatmap.png', dpi=300)
plt.show()
```

# Locating regions of high density via DBSCAN
Density-based Spatial Clustering of Applications with noise

```python
from sklearn.datasets import make_moons
X, y = make_moons(n_samples=200,
                 noise=0.05,
                 random_state=0)
plt.scatter(X[:,0], X[:,1])
```

```python
f, (ax1, ax2) = plt.subplots(1,2 , figsize=(10, 4))
km = KMeans(n_clusters=2,
           random_state=0)
y_km = km.fit_predict(X)
ax1.scatter(X[y_km==0,0],
           X[y_km==0,1],
           c="lightblue",
           marker="o", s=40,
           label="cluster 1")
ax1.scatter(X[y_km==1,0],
           X[y_km==1,1],
           c="red",
           marker="o", s=40,
           label="cluster 1")
ax1.set_title("K_mean")
ac = AgglomerativeClustering(n_clusters=2,
                            affinity="euclidean",
                            linkage='complete')
y_ac = ac.fit_predict(X)
ax2.scatter(X[y_ac==0,0],
           X[y_ac==0,1],
           c="lightblue",
           marker="o", s=40,
           label="cluster 1")
ax2.scatter(X[y_ac==1,0],
           X[y_ac==1,1],
           c="red",
           marker="o", s=40,
           label="cluster 1")
ax1.set_title("Agglomerative")
plt.legend()
```

```python
from sklearn.cluster import DBSCAN
db = DBSCAN(eps=0.3,
           min_samples=5,
           metric="euclidean")
y_db = db.fit_predict(X)
plt.figure(figsize=(10,8))
plt.scatter(X[y_db==0,0],
           X[y_db==0,1],
           c="lightblue",
           marker="o", s=40,
           label="cluster 1")
plt.scatter(X[y_db==1,0],
           X[y_db==1,1],
           c="red",
           marker="o", s=40,
           label="cluster 2")
plt.title("DBSCAN")
plt.legend()
plt.show()
```
